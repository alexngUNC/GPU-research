# Looping for 100k Preemptions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# ----- Load the data -----
dataPath = '../Data/saxpyData/loopTests/'

# First experiment
no_shared1 = pd.read_csv(dataPath+'loop100k1.csv')
shared1 = pd.read_csv(dataPath+'loop100kShared1.csv')

# rename cols
no_shared1.columns = ['start', 'end']
shared1.columns = ['start', 'end']

# interval column
no_shared1['interval'] = no_shared1['end'] - no_shared1['start']
shared1['interval'] = shared1['end'] - shared1['start']
```

## Logger Execution Intervals

```python
# ----- Plot the data -----
NUM_SAMPLES = 100000
# Plot all intervals
plt.scatter(range(1, NUM_SAMPLES), no_shared1['interval'], label='W/O shared')
plt.scatter(range(1, NUM_SAMPLES), shared1['interval'], label='With shared')

# Add labels and title
plt.xlabel('Preemption #')
plt.ylabel('Interval (ns)')
plt.title('Interval Lengths')

# Add legend
plt.legend()

# Set y-axis limits
#plt.ylim(1.01*1e3, 1.03*1e3)

#plt.yticks(range(1010, 1030))

# Show the plot
plt.show()
```

### Zoom in

```python
# ----- Plot the data -----
# Plot all intervals
plt.scatter(range(1, NUM_SAMPLES), no_shared1['interval'], label='W/O shared')
plt.scatter(range(1, NUM_SAMPLES), shared1['interval'], label='With shared')

# Add labels and title
plt.xlabel('Preemption #')
plt.ylabel('Interval (ns)')
plt.title('Interval Lengths')

# Add legend
plt.legend()

# Set y-axis limits
plt.ylim(1*1e6, 1.03*1e6)

#plt.yticks(range(1010, 1030))

# Show the plot
plt.show()
```

# Preemption and Shared Program Exeuction Intervals

```python
no_shared_ivls1 = []
shared_ivls1 = []
for i in range(0, len(shared1)-1):
  no_shared_ivls1.append(no_shared1['start'][i+1] - no_shared1['end'][i])
  shared_ivls1.append(shared1['start'][i+1] - shared1['end'][i])

# ----- Plot the data -----
# Plot all intervals
plt.scatter(range(1, NUM_SAMPLES-1), no_shared_ivls1, label='W/O shared')
plt.scatter(range(1, NUM_SAMPLES-1), shared_ivls1, label='With shared')

# Add labels and title
plt.xlabel('Preemption #')
plt.ylabel('Interval (ns)')
plt.title('Preemption and Intervals')

# Add legend
plt.legend()

# Set y-axis limits
#plt.ylim(1*1e6, 1.03*1e6)

#plt.yticks(range(1010, 1030))

# Show the plot
plt.show()
```

### Zoom In

```python
# ----- Plot the data -----
# Plot all intervals
plt.scatter(range(1, NUM_SAMPLES-1), no_shared_ivls1, label='W/O shared')
plt.scatter(range(1, NUM_SAMPLES-1), shared_ivls1, label='With shared')

# Add labels and title
plt.xlabel('Preemption #')
plt.ylabel('Interval (ns)')
plt.title('Preemption and Shared Program Execution')

# Add legend
plt.legend()

# Set y-axis limits
plt.ylim(1.25*1e6, 1.35*1e6)

#plt.yticks(range(1010, 1030))

# Show the plot
plt.show()
```

## Compare Side-by-Side

```python
# Create the subplot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))

# Scatterplot of without shared memory
ax1.scatter(range(1, NUM_SAMPLES-1), no_shared_ivls1, label='Without Shared')

# Add labels and title
ax1.set_xlabel('Preemption #')
ax1.set_ylabel('Interval')
ax1.set_title('Preemption and Shared Program Execution')

# Set y-axis limits
ax1.set_ylim(1.25*1e6, 1.35*1e6)

# Scatterplot of with shared memory
ax2.scatter(range(1, NUM_SAMPLES-1), shared_ivls1, label='With Shared', color='orange')

# Add labels and title
ax2.set_xlabel('Preemption #')
ax2.set_ylabel('Interval')
ax2.set_title('Preemption and Shared Program Execution')

# Set y-axis limits
ax2.set_ylim(1.25*1e6, 1.35*1e6)

plt.show()
```

# All Trials

```python
# ---------- Load in the rest of the data ----------
# Second trial
no_shared2 = pd.read_csv(dataPath+'loop100k2.csv')
shared2 = pd.read_csv(dataPath+'loop100kShared2.csv')

# rename cols
no_shared2.columns = ['start', 'end']
shared2.columns = ['start', 'end']

# interval column
no_shared2['interval'] = no_shared2['end'] - no_shared2['start']
shared2['interval'] = shared2['end'] - shared2['start']

# Third trial
no_shared3 = pd.read_csv(dataPath+'loop100k3.csv')
shared3 = pd.read_csv(dataPath+'loop100kShared3.csv')

# rename cols
no_shared3.columns = ['start', 'end']
shared3.columns = ['start', 'end']

# interval column
no_shared3['interval'] = no_shared3['end'] - no_shared3['start']
shared3['interval'] = shared3['end'] - shared3['start']

# Fourth trial
no_shared4 = pd.read_csv(dataPath+'loop100k4.csv')
shared4 = pd.read_csv(dataPath+'loop100kShared4.csv')

# rename cols
no_shared4.columns = ['start', 'end']
shared4.columns = ['start', 'end']

# interval column
no_shared4['interval'] = no_shared4['end'] - no_shared4['start']
shared4['interval'] = shared4['end'] - shared4['start']

# Fifth trial
no_shared5 = pd.read_csv(dataPath+'loop100k5.csv')
shared5 = pd.read_csv(dataPath+'loop100kShared5.csv')

# rename cols
no_shared5.columns = ['start', 'end']
shared5.columns = ['start', 'end']

# interval column
no_shared5['interval'] = no_shared5['end'] - no_shared5['start']
shared5['interval'] = shared5['end'] - shared5['start']

# Sixth trial
no_shared6 = pd.read_csv(dataPath+'loop100k6.csv')
shared6 = pd.read_csv(dataPath+'loop100kShared6.csv')

# rename cols
no_shared6.columns = ['start', 'end']
shared6.columns = ['start', 'end']

# interval column
no_shared6['interval'] = no_shared6['end'] - no_shared6['start']
shared6['interval'] = shared6['end'] - shared6['start']

# ---------- Accumulate all trials ----------
no_shared_ivls1 = []
shared_ivls1 = []
no_shared_ivls2 = []
shared_ivls2 = []
no_shared_ivls3 = []
shared_ivls3 = []
no_shared_ivls4 = []
shared_ivls4 = []
no_shared_ivls5 = []
shared_ivls5 = []
no_shared_ivls6 = []
shared_ivls6 = []

# Calculate start of interval i+1 - end of interval i
for i in range(0, len(shared1)-1):
  no_shared_ivls1.append(no_shared1['start'][i+1] - no_shared1['end'][i])
  shared_ivls1.append(shared1['start'][i+1] - shared1['end'][i])
  no_shared_ivls2.append(no_shared2['start'][i+1] - no_shared2['end'][i])
  shared_ivls2.append(shared2['start'][i+1] - shared2['end'][i])
  no_shared_ivls3.append(no_shared3['start'][i+1] - no_shared3['end'][i])
  shared_ivls3.append(shared3['start'][i+1] - shared3['end'][i])
  no_shared_ivls4.append(no_shared4['start'][i+1] - no_shared4['end'][i])
  shared_ivls4.append(shared4['start'][i+1] - shared4['end'][i])
  no_shared_ivls5.append(no_shared5['start'][i+1] - no_shared5['end'][i])
  shared_ivls5.append(shared5['start'][i+1] - shared5['end'][i])
  no_shared_ivls6.append(no_shared6['start'][i+1] - no_shared6['end'][i])
  shared_ivls6.append(shared6['start'][i+1] - shared6['end'][i])

# Put all intervals together
no_shared_ivls_all = no_shared_ivls1 + no_shared_ivls2 + no_shared_ivls3 + no_shared_ivls4 + no_shared_ivls5 + no_shared_ivls6
shared_ivls_all = shared_ivls1 + shared_ivls2 + shared_ivls3 + shared_ivls4 + shared_ivls5 + shared_ivls6

# ---------- Plot the data ----------
# Plot all intervals on one plot
plt.scatter(range(1, NUM_SAMPLES*6 - 11), no_shared_ivls_all, label='W/O shared')
plt.scatter(range(1, NUM_SAMPLES*6 - 11), shared_ivls_all, label='With shared')

# Add labels and title
plt.xlabel('Preemption #')
plt.ylabel('Interval (ns)')
plt.title('Preemption and Shared Program Execution')

# Add legend
plt.legend()

# Set y-axis limits
#plt.ylim(1.25*1e6, 1.35*1e6)

#plt.yticks(range(1010, 1030))

# Show the plot
plt.show()
```

## Each Interval One by One

```python
# Create subplot axes
(fig, axes) = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))

# Iterate over axes 
axes = axes.flatten()

def plot_ivl(i, no_shared_ivls, shared_ivls, axes):
  axes[i-1].scatter(np.arange(1, NUM_SAMPLES-1), no_shared_ivls, label='Without Shared')
  axes[i-1].scatter(np.arange(1, NUM_SAMPLES-1), shared_ivls, label='With Shared')
  axes[i-1].legend()
  axes[i-1].set_ylabel('Interval (ns)')
  axes[i-1].set_ylim(1.25*1e6, 1.35*1e6)

plot_ivl(1, no_shared_ivls1, shared_ivls1, axes)
plot_ivl(2, no_shared_ivls2, shared_ivls2, axes)
plot_ivl(3, no_shared_ivls3, shared_ivls3, axes)
plot_ivl(4, no_shared_ivls4, shared_ivls4, axes)
plot_ivl(5, no_shared_ivls5, shared_ivls5, axes)
plot_ivl(6, no_shared_ivls6, shared_ivls6, axes)
plt.show()
```
