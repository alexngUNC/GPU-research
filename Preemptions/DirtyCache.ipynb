# Dirty vs Clean Cache Preemption Times

```python
clean_path = 'CleanCache/xavierNoMemory1.csv'
dirty_path = 'Xavier/NoSharedData/LxavierNoSharedTwelve12th.csv'
dirty_path10 = 'xavierMap10.csv'
dirty_path100 = 'xavierMap100.csv'
dirty_path1000 = 'xavierMap1000.csv'
```

```python
from analyzer_functions import *
import numpy as np

NUM_SAMPLES = 1000000
offset = NUM_SAMPLES // 10
clean_ivls, dirty_ivls = data_loader(noSharedPath=clean_path, sharedPath=dirty_path, onlyPreemptIvls=True)
clean_ivls = np.array(clean_ivls) / 1000
dirty_ivls = np.array(dirty_ivls) / 1000

plot_side_by_side(noSharedData=clean_ivls, sharedData=dirty_ivls, NUM_SAMPLES=NUM_SAMPLES, 
    firstLabel='Clean Cache', secondLabel='Dirty Cache', lowerBound=1250, upperBound=1350, medianLines=True, offset=offset, medianOffset=10, medianImpute=True)
```

```python
dirty_ivls10 = data_loader(singlePath=dirty_path10, single=True, onlyPreemptIvls=True)
dirty_ivls10 = np.array(dirty_ivls10) / 1000

plot_side_by_side(noSharedData=clean_ivls, sharedData=dirty_ivls10, NUM_SAMPLES=NUM_SAMPLES, 
    firstLabel='Clean Cache', secondLabel='Dirty Cache', lowerBound=1250, upperBound=1350, medianLines=True, offset=offset, medianOffset=10, medianImpute=True)
```
